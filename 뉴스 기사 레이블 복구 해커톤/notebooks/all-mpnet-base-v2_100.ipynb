## Pre-processing

```python
def lower_text(text):
    return text.lower()
```

```python
# stopwords
stopwords = ['the', 'to', 'of', 'a', 'in', 'and', 'on', 'for', 'that', 'is', 'with', 'as', 'at',
             'has', 'by', 'its', 'it', 'from', 'was', 'be', 'will', 'an', 'his', 'have', 'are',
             'after', 'their', 'he', 'over', 'not', 'this', 'but', 'about', 'more', 'her', 'been',
             'they', 'were', 'would', 'had', 'could', 'can', 'you']

# мҡ”мқј лӘ©лЎқ
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

def preprocess_text(text):
    # н…ҚмҠӨнҠём—җм„ң мҳҒм–ҙ л¶Ҳмҡ©м–ҙ м ңкұ°
    text = ' '.join([word for word in text.split() if word.lower() not in stopwords])

    # н…ҚмҠӨнҠём—җм„ң '<' лҳҗлҠ” '>'лҘј нҸ¬н•Ён•ҳкұ°лӮҳ лӘЁл‘җ к°Җм§„ л¬ёмһҘмқ„ м ңкұ°н•ҳлҠ” н•ЁмҲҳ
    text = re.sub(r'&lt;[^&]*&gt;|&gt;[^&]*&lt;', '', text)

    # м •к·ң н‘ңнҳ„мӢқмқ„ мӮ¬мҡ©н•ҳм—¬ кҙ„нҳёмҷҖ н•Ёк»ҳ лӮҳмҳӨлҠ” "AP"мҷҖ "Reuters"лҘј м ңкұ°н•©лӢҲлӢӨ.
    text = re.sub(r'\(AP\)|\(Reuters\)|AP', '', text)

    # л¬ёмһҗм—ҙ ";p", "/p"лҘј лӘЁл‘җ мӮӯм ң
    text = re.sub(r'[;/]p', '', text)

    # HREFлЎң мӢңмһ‘н•ҳлҠ” нҢЁн„ҙ лӘЁл‘җ мӮӯм ң
    text = re.sub(r'HREF="[^"]*"', '', text)

    # '/'мҷҖ '\' л¬ёмһҗ лӘЁл‘җ мӮӯм ң
    text = re.sub(r'[\/\\]', ' ', text)

    # н•ҙмӢңнғңк·ё м ңкұ°
    text = re.sub(r'#\w+', '', text)

    # м •к·ң н‘ңнҳ„мӢқмқ„ мӮ¬мҡ©н•ҳм—¬ HTML м—”н„°нӢ° мҪ”л“ңлҘј м ңкұ°н•©лӢҲлӢӨ.
    text = re.sub(r'&#[0-9]+;', '', text)

    # мҲ«мһҗ м ңкұ°
    text = re.sub(r'\d+', '', text)

    # м •к·ң н‘ңнҳ„мӢқмқ„ мӮ¬мҡ©н•ҳм—¬ мҡ”мқј м ңкұ°
    text = re.sub(r"\b(" + "|".join(days_of_week) + r")\b", "", text)

    # 'html' м ңкұ°
    text = text.replace('html', '')

    # 'html short_description' м ңкұ°
    text = text.replace('html short_description', '')

    # 'short_description' м ңкұ°
    text = text.replace('short_description', '')

    # '$', '#' м ңкұ°
    text = re.sub(r'[#;\$#]', '', text)

    # кіөл°ұ л°Ҹ нҠ№мҲҳл¬ёмһҗ м ңкұ°
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df['processed_text'] = df['text'].apply(lower_text)
df['processed_text'] = df['processed_text'].apply(preprocess_text)
```

## Feature Extraction

```python
# Sentence BERT лӘЁлҚё лЎңл“ң
model = SentenceTransformer('all-mpnet-base-v2')

# н…ҚмҠӨнҠё feature м¶”м¶ң
sentence_embeddings = model.encode(df['processed_text'].tolist(), show_progress_bar=True)

from sklearn.preprocessing import MinMaxScaler

# мһ„лІ л”©мқ„ н‘ңмӨҖнҷ”
scaler = MinMaxScaler()
scaled_embeddings = scaler.fit_transform(sentence_embeddings)

# м¶”м¶ңн•ң featureлҘј лҚ°мқҙн„°н”„л Ҳмһ„м—җ м ҖмһҘ
df_embeddings = pd.DataFrame(scaled_embeddings)
```

## Clustering

```python
# Sentence BERT мһ„лІ л”©мқ„ мӮ¬мҡ©н•ҳм—¬ кө°м§‘нҷ” мҲҳн–ү
kmeans = KMeans(n_clusters=6, random_state=SEED)

df['kmeans_cluster'] = kmeans.fit_predict(scaled_embeddings)
```
