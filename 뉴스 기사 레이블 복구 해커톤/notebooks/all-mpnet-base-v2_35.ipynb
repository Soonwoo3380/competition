```python
# 제목 + 내용
df['text'] = df['title'] + '. ' + df['contents']
# df.head()
```

## Pre-processing

```python
# stopwords
stopwords = ['the', 'to', 'of', 'a', 'in', 'and', 'on', 'for', 'that', 'is', 'with', 'as', 'at',
             'has', 'by', 'its', 'it', 'from', 'was', 'be', 'will', 'an', 'his', 'have', 'are',
             'after', 'their', 'he', 'over', 'not', 'this', 'but', 'about', 'more', 'her', 'been',
             'they', 'were', 'would', 'had', 'could', 'can', 'you']

# 요일 목록
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

def preprocess_text(text):

    # 텍스트에서 영어 불용어 제거
    text = ' '.join([word for word in text.split() if word.lower() not in stopwords])

    # &lt;와 &gt;를 삭제
    text = re.sub(r'&lt;|&gt;', '', text)

    # 괄호와 함께 나오는, 그냥 나오는 "AP"와 "Reuters"를 제거
    text = re.sub(r'\(AP\)|\(Reuters\)|AP|Reuters', '', text)

    # 문자열 ";p", "/p"를 모두 삭제
    text = re.sub(r'[;/]p', '', text)

    # HREF로 시작하는 패턴 모두 삭제
    text = re.sub(r'HREF="[^"]*"', '', text)
    text = re.sub(r'href="[^"]*"', '', text)

    # '/'와 '\' 문자 모두 삭제
    text = re.sub(r'[\/\\]', '', text)

    # 해시태그 제거
    text = re.sub(r'#\w+', '', text)

    text = re.sub(r'[#;\$#]', '', text)

    # 공백 및 특수문자 제거
    text = re.sub(r'\s+', ' ', text).strip()

    # 요일 제거
    text = re.sub(r"\b(" + "|".join(days_of_week) + r")\b", '', text)

    # 월 제거
    text = re.sub(r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December|Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\b', '', text)

    # 숫자 제거
    text = re.sub(r'\d+', '', text)

    # 'short_description' 제거
    text = text.replace('short_description', '')

    # 'html' 제거
    text = text.replace('html', '')

    # 'www' 제거
    text = text.replace('www', '')

    # 'https' 제거
    text = text.replace('https', '')

    # ..과 ... 제거
    text = re.sub(r'\.{2,3}', '', text)

    return text
```

```python
df['processed_text'] = df['text'].apply(preprocess_text)
```

## Feature Extraction

```python
# Sentence BERT 모델 로드
model = SentenceTransformer('all-mpnet-base-v2')

# 텍스트 feature 추출
sentence_embeddings = model.encode(df['processed_text'].tolist(), show_progress_bar=True)

from sklearn.preprocessing import MinMaxScaler

# 임베딩을 표준화
scaler = MinMaxScaler()
scaled_embeddings = scaler.fit_transform(sentence_embeddings)

# 추출한 feature를 데이터프레임에 저장
df_embeddings = pd.DataFrame(scaled_embeddings)
```

## Clustering

```python
# Sentence BERT 임베딩을 사용하여 군집화 수행
kmeans = KMeans(n_clusters=6, random_state=SEED)

df['kmeans_cluster'] = kmeans.fit_predict(scaled_embeddings)
```
